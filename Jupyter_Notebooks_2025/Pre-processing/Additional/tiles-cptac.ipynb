```python
import cv2
import math
import os
import glob
import matplotlib.pyplot as plt
#date="13thOct"
files=[x for x in sorted(glob.glob("/home/saab/Downloads/*_main"))]
#masks=["mask"+str(x) for x in range(len(files)-1)]
no_duplicates=False
```

```python
fno=0
for file in files:
    '''if "he" in file:
        fname="images"
    else:
        fname=masks[fno]
        print(file)
        fno=fno+1'''
    fname=file.split("/")[-1]
    print(fname)
    #path=os.getcwd()+"/mltest_"+date+"/"+fname
    path=file+"/small_tiles"
    if (os.path.exists(path)==True and no_duplicates==True):
        print(os.path.exists(path)==True and no_duplicates==True)
    else:
        images=[x for x in sorted(glob.glob(file+"/large_tiles/*.png"))]
        k=0
        tile_size = (256, 256)
        offset = (256, 256)
        
        os.makedirs(path,exist_ok=True)
        for imgz in images:
            imga=cv2.imread(imgz)
            imgy=cv2.resize(imga,(3328,3328))
            img=imgy
            img_shape=img.shape
            for i in range(int(math.ceil(img_shape[0]/(offset[1] * 1.0)))):
                for j in range(int(math.ceil(img_shape[1]/(offset[0] * 1.0)))):
                    cropped_img = img[offset[1]*i:min(offset[1]*i+tile_size[1], img_shape[0]), offset[0]*j:min(offset[0]*j+tile_size[0], img_shape[1])]
            # Debugging the tiles
            # In loops, either the whole size, or the remaining size is taken
                    k=k+1
                    cv2.imwrite(os.path.join(path, f"{k:06d}"+'.png'), cropped_img)
                    if(k%20==0):
                        print("*",end=" ")
```
