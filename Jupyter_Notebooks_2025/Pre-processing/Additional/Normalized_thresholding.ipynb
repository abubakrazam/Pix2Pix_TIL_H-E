```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob
paths=sorted(glob(r"C://Users//abuba/Desktop//test_threshold//*.png"))
output=r"C://Users//abuba/Desktop//outputs//"
os.makedirs(output,exist_ok=True)
green=np.array([0,255,0],np.uint8)
```

```python
t1=120
t2=180
```

t1-> not below 100

```python
for path in paths:
    img=cv2.imread(path)
    b,g,r=cv2.split(img)
    if(max(g.reshape(-1))>t1):
        g_new=(g-np.mean(g.reshape(-1)))/(np.std(g.reshape(-1)))
        g_new=(g_new-min(g_new.reshape(-1)))/(max(g_new.reshape(-1))-min(g_new.reshape(-1)))
        g_new=np.array(g_new*255,np.uint8)
        new_image=cv2.merge([r,g_new,b])
    else:
        g_new=g
        new_image=img
    r,t=cv2.threshold(g_new,t2,255,cv2.THRESH_BINARY)
    th=cv2.merge([t,t,t])
    final_image=cv2.cvtColor(cv2.bitwise_and(th,img),cv2.COLOR_BGR2GRAY)
    cv2.imwrite(output+path.split("\\")[-1],final_image)
```
