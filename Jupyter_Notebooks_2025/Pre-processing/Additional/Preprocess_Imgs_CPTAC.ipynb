```python
import openslide
from glob import glob
import os
from matplotlib import pyplot as plt
import tiffile
import numpy as np
import PIL
import math
offset_dim=3328
tile_size=3328
```

```python
lst=glob("*_main/*.svs")
print(lst)
```

```python
for img in lst:
    pth=img.split("/")[0]+"/large_tiles"
    os.makedirs(pth,exist_ok=True)
    ig=openslide.OpenSlide(img)
    print(ig.dimensions)
    img_shape=np.asarray([ig.dimensions[0],ig.dimensions[1]])
    offset=np.asarray([offset_dim,offset_dim])
    
    k=np.array(0)
    i_condtn=int(np.ceil(img_shape[0]/(offset[1] * 1.0)).tolist())
    j_condtn=int(np.ceil(img_shape[1]/(offset[0] * 1.0)).tolist())

    x=0
    for i in range(i_condtn):
        for j in range(j_condtn):
            a=offset[1]*i
            b=offset[0]*j
            c=min(tile_size, img_shape[0]-a)
            d=min(tile_size, img_shape[1]-b)
            print("a= {} b= {} c= {} d= {}".format(a,b,c,d))
            y=ig.read_region((a, b), 0, (c, d))
            y.save(pth+"/"+f"{x:06d}"+".png")
            x=x+1
```
