```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
import os

f1=11
f2=101

path1=os.path.join(os.getcwd(),"13thOct_Imgs","blur_brightness1_"+str(f1))
path2=os.path.join(os.getcwd(),"13thOct_Imgs","blur_brightness2_"+str(f1))
path3=os.path.join(os.getcwd(),"13thOct_Imgs","blur_brightness1_"+str(f2))
path4=os.path.join(os.getcwd(),"13thOct_Imgs","blur_brightness2_"+str(f2))
path5=os.path.join(os.getcwd(),"13thOct_Imgs","he_imgs_1")

os.makedirs(path1,exist_ok=True)
os.makedirs(path2,exist_ok=True)
os.makedirs(path3,exist_ok=True)
os.makedirs(path4,exist_ok=True)
os.makedirs(path5,exist_ok=True)
```

```python
img_he=[x for x in sorted(glob(os.getcwd()+"/Preprocessed Imgs _16thSep/*_raw/he/*.png"))]
img_cd3=[x for x in sorted(glob(os.getcwd()+"/all_processed_Oct8/*/align/CD3*.png"))]
```

```python
j=0
for a,b in zip(img_he,img_cd3):
    print(a)
    background= cv2.imread(a)
    overlay = cv2.imread(b)
    background=cv2.resize(background,(3328,3328))
    overlay=cv2.resize(overlay,(3328,3328))
    overlay=cv2.cvtColor(overlay,cv2.COLOR_BGR2GRAY)
    zero=np.zeros_like(overlay)
    #ret,thresh1 = cv2.threshold(overlay,25,255,cv2.THRESH_BINARY)
    thresh1=overlay
    #Original value is 51, not 25
    temp=overlay*thresh1  # Strict thresholding to remove possible autofluoroscence sections
    kernel = np.ones((7,7),np.uint8) 
    temp=cv2.morphologyEx(temp, cv2.MORPH_CLOSE, kernel) # hole closing
    img2=cv2.merge([zero,temp,zero])
    
    img1_b1=cv2.GaussianBlur(img2,(f1,f1),7)
    img1_b2=cv2.GaussianBlur(img2,(f2,f2),2)
    a1 = cv2.addWeighted(background,0.3,img1_b1,1,0)
    a2 = cv2.addWeighted(img1_b1,1.5,background,0.5,0)
    a3 = cv2.addWeighted(background,0.3,img1_b2,1,10)
    a4 = cv2.addWeighted(img1_b2,1.5,background,0.5,0)
    cv2.imwrite(path1+"/"+f"{+j:03d}"+".png",a1)
    cv2.imwrite(path2+"/"+f"{+j:03d}"+".png",a2)
    cv2.imwrite(path3+"/"+f"{+j:03d}"+".png",a3)
    cv2.imwrite(path4+"/"+f"{+j:03d}"+".png",a4)
    
    cv2.imwrite(path5+"/"+f"{+j:03d}"+".png",background)
    
    j=j+1
```
