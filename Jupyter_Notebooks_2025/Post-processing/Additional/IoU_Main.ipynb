```python
import numpy as np
import cv2
from glob import glob
import os
import matplotlib.pyplot as plt
import pandas as pd
import tkinter
from tkinter import filedialog
from sklearn.metrics import jaccard_score
tkinter.Tk().withdraw() # prevents an empty tkinter window from appearing
```

```python
path1=filedialog.askdirectory(title='path for ground truth images')
path2=filedialog.askdirectory(title='path for predicted images')
output_imgs=sorted(glob(path1+"/*.png"))
input_imgs=sorted(glob(path2+"/*.png"))
```

```python
print(input_imgs[:2])
print(output_imgs[:2])
```

```python
#Simple check to ensure same non-zero number of images going in
len(input_imgs)==len(output_imgs)!=0
```

```python
lst=[]
l_p=[]
l_tf=[]
for i1,i2 in zip(input_imgs,output_imgs):
    
    img1=cv2.imread(i1,0)
    img2=cv2.imread(i2,0)
    i_1=img1.reshape(-1)/(255)
    i_2=img2.reshape(-1)/(255)
    if(np.count_nonzero(i_1)>100):
        print(i1)
        '''
    if(len(np.unique(i_1))!=1 and len(np.unique(i_2))!=1):
        l_p.append(i1.split("/")[-1])
        j=jaccard_score(i_1,i_2 , average="weighted")
        lst.append(j)
        if(j>0.5):
            l_tf.append(1)
        else:
            l_tf.append(0)
        '''
    
        l_p.append(i1.split("/")[-1])
        values,counts=np.unique(i_1*2+i_2,return_counts=True)
        if(len(values)>2):
            j=counts[len(values)-1]/(counts[len(values)-2]+counts[len(values)-1])
        
        elif(values[-1]==2):
            j=0
        else:
            j=1
        lst.append(j)
        
    #l_tf.append(j)
        if(j>0.5):
            l_tf.append(1)
        else:
            l_tf.append(0)
    
```

```python
lst=[]
l_p=[]
l_tf=[]
path4="/home/saab/Desktop"
path5="/home/saab/Desktop/SGH_IMCB_NTU_Main/Processed_Data/Preprocessed_Data/Serial_Slide/15thOct/bn_tiles"
file_paths=pd.read_csv(path4+"/file_paths.csv")
for i in range(len(input_imgs)):
    
    img1=cv2.imread(input_imgs[i],0)
    img2=cv2.imread(output_imgs[i],0)
    img3=cv2.bitwise_and(img1,img2)
    bn=cv2.imread(path5+"/"+file_paths.loc[i][1],0)
    img4=(img3/255)*bn
    img5=(img2/255)*bn
    
    l_p.append(input_imgs[i].split("/")[-1])
    v1,c1=np.unique(img5,return_counts=True)
    v2,c2=np.unique(img4,return_counts=True)
    lst.append(len(v1))
    l_tf.append(len(set(v1)&set(v2)))
a1=np.average(lst)
a2=np.average(l_tf)
lst.append(a1)
l_tf.append(a2)
l_p.append("Total")
df=pd.DataFrame({'name': l_p,'GT':lst,'Pred':l_tf})
path3=filedialog.askdirectory(title='path for csv')
df.to_csv(path3+"/test_IoU_1510_serial_cell.csv",index=False)
```
